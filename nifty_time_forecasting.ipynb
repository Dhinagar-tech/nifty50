```python
import pandas as pd

# Load the CSV file
file_path = 'NIFTY50.csv'
df = pd.read_csv(file_path)

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Check for missing values in the dataset
missing_values = df.isnull().sum()

# Fill missing values 
df['Open'] = df['Open'].fillna(method='ffill')
df['High'] = df['High'].fillna(method='ffill')
df['Low'] = df['Low'].fillna(method='ffill')
df['Close'] = df['Close'].fillna(method='ffill')
df['Adj Close'] = df['Adj Close'].fillna(method='ffill')
df['Volume'] = df['Volume'].fillna(0)

# Summary statistics
summary_stats = df.describe()

# Checking for missing values after handling them 
missing_values_after = df.isnull().sum()

# Save the cleaned data to a new CSV file
output_path = 'NIFTY50_cleaned.csv'
df.to_csv(output_path, index=False)

# Display results
print("Missing Values Before Handling:\n", missing_values)
print("\nSummary Statistics:\n", summary_stats)
print("\nMissing Values After Handling:\n", missing_values_after)
print(f"\nCleaned data has been saved to {output_path}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

df = pd.read_csv('NIFTY50_cleaned.csv')
print(df.head())

print(df.info())

df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)


```

```python
# Data Preparation: Creating lag features and splitting the data into train-test sets

# Create lag features (e.g., previous day's 'Close' price)
df['Lag_1'] = df['Close'].shift(1)
df['Lag_2'] = df['Close'].shift(2)

# Drop any rows with missing values due to the lag
df.dropna(inplace=True)

# Define target and features
X = df[['Lag_1', 'Lag_2']]  # Using lagged prices as features
y = df['Close']  # Target is the 'Close' price

# Split the data into train and test sets (80% train, 20% test)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Verify the shapes of the datasets
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Import necessary libraries for machine learning and evaluation
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Initialize and train the RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_rf = rf_model.predict(X_test)

# Calculate evaluation metrics
mse_rf = mean_squared_error(y_test, y_pred_rf)
mae_rf = mean_absolute_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

mse_rf, mae_rf, r2_rf
```

LSTM

```python
# Import necessary libraries for LSTM and data scaling
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Scaling the 'Close' prices between 0 and 1 for LSTM input
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df[['Close']])

# Function to create dataset with time steps (lag features for LSTM)
def create_dataset(data, time_step=1):
    X, Y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        Y.append(data[i + time_step, 0])
    return np.array(X), np.array(Y)

# Define the time step for LSTM
time_step = 10
X_lstm, y_lstm = create_dataset(scaled_data, time_step)

# Reshape X for LSTM input [samples, time_steps, features]
X_lstm = X_lstm.reshape((X_lstm.shape[0], X_lstm.shape[1], 1))

# Split the data into training and testing sets (80% train, 20% test)
train_size_lstm = int(len(X_lstm) * 0.8)
X_train_lstm, X_test_lstm = X_lstm[:train_size_lstm], X_lstm[train_size_lstm:]
y_train_lstm, y_test_lstm = y_lstm[:train_size_lstm], y_lstm[train_size_lstm:]

# Build the LSTM model
lstm_model = Sequential()
lstm_model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
lstm_model.add(LSTM(50, return_sequences=False))
lstm_model.add(Dense(25))
lstm_model.add(Dense(1))

# Compile the model
lstm_model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
lstm_model.fit(X_train_lstm, y_train_lstm, epochs=50, batch_size=64, verbose=0)

# Make predictions
lstm_predictions = lstm_model.predict(X_test_lstm)
lstm_predictions = scaler.inverse_transform(lstm_predictions)

# Invert the scaling for actual y_test as well
y_test_lstm = y_test_lstm.reshape(-1, 1)
y_test_lstm_actual = scaler.inverse_transform(y_test_lstm)

# Calculate LSTM evaluation metrics
mse_lstm = mean_squared_error(y_test_lstm_actual, lstm_predictions)
mae_lstm = mean_absolute_error(y_test_lstm_actual, lstm_predictions)
r2_lstm = r2_score(y_test_lstm_actual, lstm_predictions)

# Print evaluation metrics
print(f"LSTM Mean Squared Error (MSE): {mse_lstm}")
print(f"LSTM Mean Absolute Error (MAE): {mae_lstm}")
print(f"LSTM R² Score: {r2_lstm}")

import matplotlib.pyplot as plt

# Plotting LSTM predictions vs actual values
plt.figure(figsize=(12, 6))
plt.plot(y_test_lstm_actual, label="Actual Prices", color="blue")
plt.plot(lstm_predictions, label="LSTM Predicted Prices", color="red")
plt.title("LSTM Model - Actual vs Predicted Prices")
plt.xlabel("Time Steps")
plt.ylabel("Close Price")
plt.legend()
plt.grid(True)
plt.show()



```

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Build and fit the ARIMA model (p=5, d=1, q=0 as an example, you can adjust these)
arima_model = ARIMA(df['Close'], order=(5, 1, 0))
arima_fit = arima_model.fit()

# Forecast the test data period
arima_predictions = arima_fit.forecast(steps=len(y_test))
arima_test = df['Close'][-len(y_test):]  # Actual test set

# Calculate ARIMA evaluation metrics
mse_arima = mean_squared_error(arima_test, arima_predictions)
mae_arima = mean_absolute_error(arima_test, arima_predictions)
r2_arima = r2_score(arima_test, arima_predictions)

# Print ARIMA evaluation metrics
print(f"ARIMA Mean Squared Error (MSE): {mse_arima}")
print(f"ARIMA Mean Absolute Error (MAE): {mae_arima}")
print(f"ARIMA R² Score: {r2_arima}")

# Plotting actual vs predicted
plt.figure(figsize=(10,6))
plt.plot(arima_test.index, arima_test, label='Actual')
plt.plot(arima_test.index, arima_predictions, label='ARIMA Predicted')
plt.legend()
plt.show()
```

```python
# Import necessary libraries for Decision Tree and data scaling
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (adjust the file path accordingly)
df = pd.read_csv('NIFTY50_cleaned.csv')  # Replace 'your_data.csv' with your actual file path

# Scaling the 'Close' prices between 0 and 1 for Decision Tree input
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df[['Close']])

# Function to create dataset with time steps (lag features for Decision Tree)
def create_dataset(data, time_step=1):
    X, Y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        Y.append(data[i + time_step, 0])
    return np.array(X), np.array(Y)

# Define the time step
time_step = 10
X_dt, y_dt = create_dataset(scaled_data, time_step)

# Reshape X for Decision Tree input [samples, features]
X_dt = X_dt.reshape(X_dt.shape[0], X_dt.shape[1])

# Split the data into training and testing sets (80% train, 20% test)
train_size_dt = int(len(X_dt) * 0.8)
X_train_dt, X_test_dt = X_dt[:train_size_dt], X_dt[train_size_dt:]
y_train_dt, y_test_dt = y_dt[:train_size_dt], y_dt[train_size_dt:]

# Build the Decision Tree model
dt_model = DecisionTreeRegressor()

# Train the model
dt_model.fit(X_train_dt, y_train_dt)

# Make predictions
dt_predictions = dt_model.predict(X_test_dt)

# Invert the scaling for predicted and actual values
dt_predictions_actual = scaler.inverse_transform(dt_predictions.reshape(-1, 1))
y_test_dt_actual = y_test_dt.reshape(-1, 1)
y_test_dt_actual = scaler.inverse_transform(y_test_dt_actual)

# Calculate Decision Tree evaluation metrics
mse_dt = mean_squared_error(y_test_dt_actual, dt_predictions_actual)
mae_dt = mean_absolute_error(y_test_dt_actual, dt_predictions_actual)
r2_dt = r2_score(y_test_dt_actual, dt_predictions_actual)

# Print evaluation metrics
print(f"Decision Tree Mean Squared Error (MSE): {mse_dt}")
print(f"Decision Tree Mean Absolute Error (MAE): {mae_dt}")
print(f"Decision Tree R² Score: {r2_dt}")

# Plotting Decision Tree predictions vs actual values
plt.figure(figsize=(12, 6))
plt.plot(y_test_dt_actual, label="Actual Prices", color="blue")
plt.plot(dt_predictions_actual, label="Decision Tree Predicted Prices", color="red")
plt.title("Decision Tree Model - Actual vs Predicted Prices")
plt.xlabel("Time Steps")
plt.ylabel("Close Price")
plt.legend()
plt.grid(True)
plt.show()
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

# Define the time step for lag features
time_step = 10

# Create lag features (similar to the LSTM input)
def create_lag_features(data, time_step=1):
    X, Y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        Y.append(data[i + time_step, 0])
    return np.array(X), np.array(Y)

# Create dataset with lag features
X_rf, y_rf = create_lag_features(scaled_data, time_step)

# Split the data into training and testing sets (80% train, 20% test)
X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(
    X_rf, y_rf, test_size=0.2, random_state=42, shuffle=False
)

# Initialize Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the Random Forest Regressor
rf_model.fit(X_train_rf, y_train_rf)

# Make predictions
rf_predictions = rf_model.predict(X_test_rf)

# Rescale predictions and test data back to the original scale
rf_predictions = rf_predictions.reshape(-1, 1)
rf_predictions_actual = scaler.inverse_transform(rf_predictions)

y_test_rf = y_test_rf.reshape(-1, 1)
y_test_rf_actual = scaler.inverse_transform(y_test_rf)

# Calculate Random Forest evaluation metrics
mse_rf = mean_squared_error(y_test_rf_actual, rf_predictions_actual)
mae_rf = mean_absolute_error(y_test_rf_actual, rf_predictions_actual)
r2_rf = r2_score(y_test_rf_actual, rf_predictions_actual)

# Print evaluation metrics
print(f"Random Forest Mean Squared Error (MSE): {mse_rf}")
print(f"Random Forest Mean Absolute Error (MAE): {mae_rf}")
print(f"Random Forest R² Score: {r2_rf}")

# Plotting Random Forest predictions vs actual values
plt.figure(figsize=(12, 6))
plt.plot(y_test_rf_actual, label="Actual Prices", color="blue")
plt.plot(rf_predictions_actual, label="RF Predicted Prices", color="green")
plt.title("Random Forest Model - Actual vs Predicted Prices")
plt.xlabel("Time Steps")
plt.ylabel("Close Price")
plt.legend()
plt.grid(True)
plt.show()
```
